```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("C:\\Users\\dnnxl\\Documents\\GitHub\\llm-param-search\\dataset\\FEINA_test.csv")  # ‚Üê change to your file path

# Random 80/20 split
train_df, test_df = train_test_split(
    df,
    test_size=0.2,      # 20% test
    random_state=42,    # makes the split reproducible
    shuffle=True        # ensures random shuffling
)

print("Train size:", len(train_df))
print("Test size:", len(test_df))

train_df.to_csv("C:\\Users\\dnnxl\\Documents\\GitHub\\llm-param-search\\dataset\\FEINA_test_split_train.csv", index=False)
test_df.to_csv("C:\\Users\\dnnxl\\Documents\\GitHub\\llm-param-search\\dataset\\FEINA_test_split_test.csv", index=False)
```
